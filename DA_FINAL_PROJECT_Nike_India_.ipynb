```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://raw.githubusercontent.com/rojer2720/Main_Project/refs/heads/main/Nike_Sales_Uncleaned.csv'

df = pd.read_csv(url)
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Order_ID as string (identifier, not numeric)
df['Order_ID'] = df['Order_ID'].astype(str)

# Parse 3 mixed date formats: 2024-03-09 | 04-10-2024 | 2024/09/12
def parse_date(d):
    if pd.isna(d):
        return pd.NaT
    for fmt in ('%Y-%m-%d', '%d-%m-%Y', '%Y/%m/%d'):
        try:
            return pd.to_datetime(d, format=fmt)
        except:
            pass
    return pd.NaT

df['Order_Date'] = df['Order_Date'].apply(parse_date)

print('Order_Date dtype :', df['Order_Date'].dtype)
print('Order_Date nulls :', df['Order_Date'].isna().sum())
```

```python
#cities name corrected
print('Cities before fix:', df['Region'].unique())

city_map = {
    'bengaluru' : 'Bengaluru',
    'Bangalore'  : 'Bengaluru',
    'Hyd'        : 'Hyderabad',
    'hyderbad'   : 'Hyderabad',
    'Hyderabad'  : 'Hyderabad',
    'Mumbai'     : 'Mumbai',
    'Pune'       : 'Pune',
    'Delhi'      : 'Delhi',
    'Kolkata'    : 'Kolkata',
}
df['Region'] = df['Region'].map(city_map)
print('Cities after fix :', df['Region'].unique())
```

```python
#Handle Duplicates
print('Rows before:', len(df))
print('Duplicate order_id rows :', df.duplicated('Order_ID').sum())

df.drop_duplicates(subset='Order_ID', keep='first', inplace=True)
df.reset_index(drop=True, inplace=True)

print('Rows after :', len(df))
```

```python

print("Duplicate Rows :", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
# Standardise to uppercase string
df['Size'] = df['Size'].astype(str).str.strip().str.upper()
df['Size'] = df['Size'].replace('NAN', np.nan)

print('Unique sizes before mapping:', sorted(df['Size'].dropna().unique()))
```

```python
# Map numeric shoe sizes to apparel sizes
shoe_to_apparel = {'6': 'S',  '7': 'S','8': 'M',  '9': 'M','10': 'L', '11': 'L','12': 'XL'}
df['Size'] = df['Size'].replace(shoe_to_apparel)

print('Unique sizes after mapping:', sorted(df['Size'].dropna().unique()))
```

```python
# filling missing size using mode per product_name
def fill_mode(series):
    m = series.mode()
    return series.fillna(m[0]) if len(m) > 0 else series

df['Size'] = df.groupby('Product_Name')['Size'].transform(fill_mode)

print('size nulls after imputation:', df['Size'].isna().sum())
print('size value counts:')
print(df['Size'].value_counts())
```

```python
# Units_Sold < 0 → returns / errors → NaN
print('Units_Sold < 0 :', (df['Units_Sold'] < 0).sum())
df.loc[df['Units_Sold'] < 0, 'Units_Sold'] = np.nan

# Discount_Applied > 1 → above 100%, impossible → NaN
print('Discount_Applied > 1   :', (df['Discount_Applied'] > 1).sum())
df.loc[df['Discount_Applied'] > 1, 'Discount_Applied'] = np.nan

# Revenue <= 0 → data error → NaN (covers zeros AND negatives)
print('Revenue <= 0   :', (df['Revenue'] <= 0).sum())
df.loc[df['Revenue'] <= 0, 'Revenue'] = np.nan
```

```python
# Units_Sold → median per Product_Line
df['Units_Sold'] = df.groupby('Product_Line')['Units_Sold'] \
                     .transform(lambda x: x.fillna(x.median()))

# MRP → median per Product_Name
df['MRP'] = df.groupby('Product_Name')['MRP'] \
               .transform(lambda x: x.fillna(x.median()))

# Discount_Applied → median per Product_Line
df['Discount_Applied'] = df.groupby('Product_Line')['Discount_Applied'] \
                   .transform(lambda x: x.fillna(x.median()))

# Order_Date → mode per Sales_Channel
df['Order_Date'] = df.groupby('Sales_Channel')['Order_Date'] \
                     .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else pd.NaT))

# Safety net: any still-missing numerics → global median
for col in ['Units_Sold', 'MRP', 'Discount_Applied']:
    df[col] = df[col].fillna(df[col].median())

print('Missing after imputation:')
print(df.isnull().sum())
```

```python
# Recalculate revenue for all missing rows
revenue_mask = df['Revenue'].isna()
df.loc[revenue_mask, 'Revenue'] = (
    df.loc[revenue_mask, 'Units_Sold'] *
    df.loc[revenue_mask, 'MRP'] *
    (1 - df.loc[revenue_mask, 'Discount_Applied'])
).round(2)

# Convert Units_Sold to integer
df['Units_Sold'] = df['Units_Sold'].astype(int)

print('Revenue nulls :', df['Revenue'].isna().sum())
print('Revenue == 0  :', (df['Revenue'] == 0).sum())
print('Revenue < 0   :', (df['Revenue'] < 0).sum())
```

```python
def winsorize_iqr(df, col, factor=1.5):
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - factor * IQR
    upper = Q3 + factor * IQR
    n_out = ((df[col] < lower) | (df[col] > upper)).sum()
    df[col] = df[col].clip(lower=lower, upper=upper)
    print(f'{col:15s} | lower={lower:10.2f} | upper={upper:10.2f} | capped={n_out}')

print(f'{"Column":15s} | {"Lower fence":>14} | {"Upper fence":>14} | Capped')
print('-' * 60)
for col in ['MRP', 'Revenue', 'Profit']:
    winsorize_iqr(df, col)
```
